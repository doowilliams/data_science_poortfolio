### Let 's start using a Histogram

```python
df.hist(column = 'alcohol', bins = 15) # bins = 5 to more. to get  more graphic
```

```python
# let get a bit more advanced and learn to customize your plots
df.hist(column = 'alcohol', bins = 10 ,grid = False, figsize =(10, 6),color ='green')
```

```python
import seaborn as sns
sns.distplot(df['alcohol'], bins = 25, kde = False )
```

```python
sns.distplot(df['alcohol'], bins = 25,kde = True, color= 'brown')
```

```python
df['alcohol'].value_counts().head()
```

```python
# lets start with labeling our axis

import  matplotlib.pyplot as plt

_= sns.distplot(df.alcohol)
_= plt.xlabel('Alcohol perecentage')
_= plt.ylabel('count')
_=plt.title('Alcohol Content')
plt.show()
```

```python
# use sns.set_style() to control styling

sns.set_style('dark')
sns.distplot(df.alcohol, bins = 15)
```

### Scatter plots

```python
sns.scatterplot(x = 'alcohol', y = 'fixed acidity', data=df)
```

```python
# scatterplot arguments

sns.scatterplot(x = 'alcohol', y = 'chlorides', hue = 'type',data = df)
```

```python
sns.lmplot(x = 'alcohol',y = 'chlorides',hue = 'type',data = df)
```

```python
# Bar

# Enlarge the plot

plt.figure(figsize=(12,8))
sns.countplot(x = 'quality',data =df, hue = 'type')
```
