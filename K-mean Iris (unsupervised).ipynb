## Adjusting the dataset for unsupervised learning....using either iloc or loc

```python
features=df.iloc[:,0:4]
```

```python
features.head()
```

### implementing the K-means Clustering

```python
from sklearn.cluster import KMeans

kmns =KMeans(n_clusters=99)  # 99 CLUSTERS ???
```

### why 99??   Because i dont know the right amount of labels.Dont worry, there is a solution for it.**Finding the best amount of clusters to get most accurate results (K-means)** i will use ELBOW RULE which is basically looking for  a plot line that respectively has a slope nearest to 90 degrees compared to y axis and be smallest as possible(yes, it looks like an elbow)

```python
kmns.fit(features)
```

```python
labels=kmns.predict(features)

labels
```

```python
ctn =kmns.cluster_centers_

ctn
```

```python
plt.scatter(features.iloc[:,2],features.iloc[:,3], c=labels)

plt.scatter(ctn[:,2],ctn[:,3],marker='o',color='red',s=120)

plt.xlabel('Petal_Lenth(cm)')
plt.ylabel('Petal_Width(cm)')
plt.show()
```
